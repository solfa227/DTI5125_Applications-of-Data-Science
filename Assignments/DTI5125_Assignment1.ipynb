```python
# Import the libraries that we will use to help create the train and test sets
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import re
import os
import csv 
import random
from random import seed
from random import random
from random import randrange

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
# Fetch the filenames from current directory
files = [file for file in os.listdir('.') if os.path.isfile(file) and file.endswith('.txt')]
```

```python
# Prepare the data in a csv file dataframe
FinalSec=[]
FirstFile='Yes'
for file in files: 
    # Read each text file as a dataframe form the main folder
    df = pd.read_csv(file, header=None, error_bad_lines=False)
    #df = pd.read_csv(file, header=None, on_bad_lines='skip')
    df.head(5)
    # Select the first column in the DataFrame
    df_col = df.iloc[:,0]
    # Data clean up
    df_lower = df_col.str.lower()
    df_NoComma = df_lower.str.replace(',', '')
    # Remove stop-words and garbage characters 
    df_NoStop = df_NoComma.apply(lambda x: [item for item in x.split() if item not in stop])
    df_final = df_NoStop
    #print(df_final)
    len_rows = len(df_final)
    # Find the number of words in cleaned up df
    word_count = 0
    word_array = []
    sections = {}
    part = 1
    for i in range(len_rows):
        result = re.findall(r"[a-zA-z]+", str(df_final[i]))
        if len(result) + word_count < 100:
            word_count = len(result) + word_count
            word_array = result + word_array
        else: 
            remainder = 100 - word_count
            word_array = word_array + result[:remainder]
            sections[part] = word_array
            word_count = 0
            word_array = []
            part += 1
    counter = 1
    seed(1)
    while counter < 201:
        rand = randrange(len(sections)-1)+1
        xx = sections.get(rand)
        if FirstFile=='Yes':
            FinalSec=FinalSec+[[xx]]
        else:
            FinalSec[counter-1]=FinalSec[counter-1]+[xx]
        counter +=1
    FirstFile='No'
```

```python
# write output files into a csv file in designated sub-folder for that file
book_file=open("output_file.csv","w",newline="")
writer=csv.writer(book_file)
for newcounter in range(200):
    writer.writerow(FinalSec[newcounter])
book_file.close()
```
